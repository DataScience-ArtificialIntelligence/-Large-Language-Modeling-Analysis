```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("facebook/m2m100_418M")
model = AutoModelForSeq2SeqLM.from_pretrained("facebook/m2m100_418M")
```

```python
import pandas as pd
import numpy as np
import csv
```

```python
df=pd.read_csv('./Bhagwad_Gita.csv')
```

```python
df.head()
```

```python
Shlokas=df['Shloka'].values
Hindi=df['HinMeaning'].values
English=df['EngMeaning'].values
```

```python
import time

start_time = time.time()
translatedHinEng = []

for shloka in Hindi[:100]:
    encoded_en = tokenizer(shloka, return_tensors="pt", padding=True, truncation=True)
    generated_tokens = model.generate(**encoded_en, forced_bos_token_id=tokenizer.get_lang_id("en"))
    translation = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)[0]
    print(f"Hindi: {shloka}\nEnglish: {translation}\n")
    translatedHinEng.append(translation)

end_time = time.time()
total_time = end_time - start_time

print(f"Total time taken for translating first 100 lines: {total_time:.2f} seconds")
```

```python
file_path = '/content/data1.csv'
with open(file_path, 'w', newline="") as file:
    writer = csv.writer(file)
    writer.writerows(translatedHinEng)
```

```python
references = [[eng] for eng in English]  # Create reference list of lists
candidates = translatedHinEng  # Model translations
```

```python
import numpy as np
from nltk.translate.bleu_score import sentence_bleu

bleu_scores = []
for i in range(len(references)):
    bleu_scores.append(sentence_bleu(references[i], candidates[i], weights=(0.25, 0.25, 0.25, 0.25)))

bleu_score = np.mean(bleu_scores)
print(f"BLEU score: {bleu_score:.4f}")
```
