```python
from transformers import MBartForConditionalGeneration, MBart50TokenizerFast
```

```python
model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-one-to-many-mmt")

tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-one-to-many-mmt", src_lang="en_XX")
```

```python
import pandas as pd
```

```python
df=pd.read_csv('./Bhagwad_Gita.csv')
```

```python
df.head()
```

```python
Shlokas=df['Shloka'].values
Hindi=df['HinMeaning'].values
English=df['EngMeaning'].values
```

```python
import numpy as np
import csv
```

```python
translatedEngHin1 = []
for shloka in English[:400]:
    model_inputs = tokenizer(shloka, return_tensors="pt")
    generated_tokens = model.generate(
        **model_inputs,
        forced_bos_token_id=tokenizer.lang_code_to_id["hi_IN"]
    )
    translation = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
    print(f"English: {shloka}\nHindi: {translation}\n")
    translatedEngHin1.append(translation)
```

```python
file_path = '/content/data1.csv'
with open(file_path, 'w', newline="") as file:
    writer = csv.writer(file)
    writer.writerows(translatedEngHin1)
```

```python
translatedEngHin2 = []
cnt = 0
for shloka in English[400:]:
    model_inputs = tokenizer(shloka, return_tensors="pt")
    generated_tokens = model.generate(
        **model_inputs,
        forced_bos_token_id=tokenizer.lang_code_to_id["hi_IN"]
    )
    translation = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True)
    cnt = cnt + 1
    print(cnt)
    print(f"English: {shloka}\nHindi: {translation}\n")
    translatedEngHin2.append(translation)

```

```python
file_path = '/content/data2.csv'
with open(file_path, 'w', newline="") as file:
    writer = csv.writer(file)
    writer.writerows(translatedEngHin2)
```

```python
import re

for i in range(len(references)):
    references[i] = re.sub(r'।।\d+\.\d+।।', '', str(references[i]))
    candidates[i] = re.sub(r'\d+\.\d+', '', str(candidates[i]))

for i in range(len(references)):
    print("Reference:", references[i])
    print("Candidate:", candidates[i])
    print("----------------------------")
```

```python
import numpy as np
from nltk.translate.bleu_score import sentence_bleu
references = [[hindi] for hindi in Hindi]  # Create reference list of lists
candidates = translatedEngHin  # Model translations

bleu_scores = []
for i in range(len(references)):
    bleu_scores.append(sentence_bleu(references[i], candidates[i], weights=(0.25, 0.25, 0.25, 0.25)))

bleu_score = np.mean(bleu_scores)
print(f"BLEU score: {bleu_score:.4f}")
```
